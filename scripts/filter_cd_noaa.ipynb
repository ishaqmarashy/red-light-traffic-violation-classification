```python
import pandas as pd
from datetime import datetime, timedelta
import os
csv_filename1 = "../data/Climate_Data/noaa/3547240.csv"


df1 = pd.read_csv(csv_filename1)
```

```python
combined_df = df1
combined_df
```

```python
combined_df.isna().sum()
```

```python
threshold = 50
filtered_df = combined_df.dropna(axis=1, thresh=len(combined_df) - threshold)

combined_df.dropna(axis=1, thresh=len(combined_df) - threshold, inplace=True)
combined_df.drop(columns=['STATION','NAME','ELEVATION','AWND_ATTRIBUTES','TMAX_ATTRIBUTES','TMIN_ATTRIBUTES','WSF2_ATTRIBUTES','PGTM_ATTRIBUTES','PRCP_ATTRIBUTES','WDF2_ATTRIBUTES','WDF5_ATTRIBUTES','WSF5_ATTRIBUTES'],inplace=True)
```

```python
combined_df['DATE'] = pd.to_datetime(combined_df['DATE'])
combined_df['DATE'] = combined_df['DATE'].dt.round('H')
combined_df
```

```python
na_counts = combined_df.isna().sum()
for column, count in na_counts.items():
    print(f"{column}: {count}")
```

```python
df=combined_df
```

```python
start_date = datetime(2012, 1, 1)
end_date = datetime(2023, 9, 30) 
expected_dates = pd.date_range(start=start_date, end=end_date, freq='H')
reference_df = pd.DataFrame({'DATE': expected_dates})

df = pd.merge(reference_df, df, on='DATE', how='outer')

df = df.set_index('DATE').ffill().bfill().reset_index()

df
```

```python
na_counts = df.isna().sum()
for column, count in na_counts.items():
    print(f"{column}: {count}")
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
df['DATE'] = df['DATE'].dt.round('H')

numeric_columns = ["LATITUDE", "LONGITUDE", "AWND",
                   "PGTM", "PRCP", "TMAX",
                   "WSF2", "WSF5",
                   "TMIN", "WDF2", "WDF5"]
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

df = df.set_index('DATE').ffill().bfill().reset_index()
result = df.groupby(['DATE']).agg({**{col: 'mean' for col in numeric_columns}}).reset_index()
result
```

```python
result.isna().sum()
```

```python
csv_filename = '../data/Climate_Data/noaa/3547240_cleaned.csv'
result.to_csv(csv_filename, index=False)
```
